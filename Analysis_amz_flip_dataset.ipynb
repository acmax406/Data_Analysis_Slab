```python
#using hamming distance to recognise the similar entry
def hammingDist(str1, str2):
    i = 0
    count = 0
    if len(str(str1))>len(str(str2)):
        while i < len(str(str2)):
            if(str1[i] != str2[i]):
                count += 1
            i += 1
        return count
    else:
        while i < len(str(str1)):
            if(str1[i] != str2[i]):
                count += 1
            i += 1
        return count
        
```

```python
#searching the similar product specifications in the amazon dataset
def amz_data(text,df_amz):
    list_amz=[]
    for j in df_amz['product_specifications']:
        dist_amz = hammingDist(text,j)
        list_amz.append(dist_amz)
    df_amz['values'] = list_amz
    m = df_amz.loc[df_amz['values'] == min(df_amz['values'])]
    return m  
```

```python
def find_indata_name(text,df_flip,df_amz):
    list_flip= [] #store the hamming dist. of each of the product nmae
    df_flip = df_flip[df_flip.product_name.str.len() >= len(text)].copy()

    for i in df_flip['product_name']:
        dist = hammingDist(i,text)
#         dist = dist/(len(text)+(len(i))) #normalising the hamming dist        
        list_flip.append(dist) #appending each of the hamming dist
        
    df_flip['values'] = list_flip #adding it to the dataframe
    
    m=df_flip.loc[df_flip['values'] == min(df_flip['values'])] #taking out the rows which have the list hamming dist
   # creating empty list to store the value of amazon dataset
    product_name_amz=[]
    retail_price_amz =[]
    discount_price_amz=[]
    
    #traverisng the obtained data from the amz_data and storing in the different columns
    for i in range(len(m)):
        data= amz_data(m['product_specifications'].iloc[i],df_amz)
        
        product_name_amz.append(data['product_name'].iloc[0])
        retail_price_amz.append(data['retail_price'].iloc[0])
        discount_price_amz.append(data['discounted_price'].iloc[0])
    
    #creating the final dataframe
    df_result_flip = pd.DataFrame({
        "Product name in Flipkart":m['product_name'],
        "Retail Price in Flipkart":m['retail_price'],
        "Discounted Price in Flipkart":m['discounted_price'],
        "Product name in amazon":product_name_amz,
        "Retail Price in amazon":retail_price_amz,
        "Discounted Price in amazon":discount_price_amz
        
    })
    return df_result_flip

```
